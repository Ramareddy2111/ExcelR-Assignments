```python
WINE=wine.values
WINE
```

```python
wine_normal=scale(WINE)
wine_normal
```

```python
pca=PCA()
pca_values=pca.fit_transform(wine_normal)
pca_values
```

```python
pca=PCA(n_components=3)
pca_values=pca.fit_transform(wine_normal)
pca_values
```

```python
var=pca.explained_variance_ratio_
var
```

```python
var1=np.cumsum(np.round(var,decimals=4)*100)
var1
```

```python
pca.components_
```

```python
plt.plot(var1, color='green')
plt.show()
```

```python
pca_values[:,0:3]
```

**Clustering using first three principal component**

```python
dendrogram=sch.dendrogram(sch.linkage(pca_values, method='complete'))
```

```python
hc=AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='single')
```

```python
y_hc=hc.fit_predict(pca_values)
Clusters=pd.DataFrame(y_hc, columns=['Clusters'])
Clusters
```

```python
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i, random_state=0)
  kmeans.fit(pca_values)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
clusters_new=KMeans(2, random_state=42)
clusters_new.fit(pca_values)
```

```python
clusters_new.labels_
```

**Perforing cluster analysis on all data**

```python
def norm_func(i):
  x=(i-i.min())/(i.max()-i.min())
  return(x)
```

```python
df_norm=norm_func(wine.iloc[:,1:])
```

```python
dendrogram=sch.dendrogram(sch.linkage(df_norm, method='single'))
```

```python
hc=AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='single')
```

```python
y_hc=hc.fit_predict(df_norm)
Clusters=pd.DataFrame(y_hc, columns=['Clusters'])
Clusters
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_wine_df=scaler.fit_transform(wine.iloc[:,1:])
```

```python
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i, random_state=0)
  kmeans.fit(scaled_wine_df)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```
