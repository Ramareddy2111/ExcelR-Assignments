```python
plt.subplots(figsize=(8,6))
plt.subplot(121)
plt.boxplot(df['Unit A'])
plt.title('Unit A')
plt.subplot(122)
plt.boxplot(df['Unit B'])
plt.title('Unit B')
plt.show()
```

```python
plt.figure(figsize=(5,5))
sm.qqplot(df["Unit A"], line='q')
plt.title('Unit A')
sm.qqplot(df["Unit B"], line='q')
plt.title('Unit B')
plt.show()
```

```python
p_value= stats.ttest_ind(df['Unit A'],df['Unit B'])
p_value
```

```python
print('p_value',p_value[1])
```
